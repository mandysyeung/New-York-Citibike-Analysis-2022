```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_cleaned = pd.read_csv('cleaned_citibike_daily_rides.csv')
```

```python
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'])
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))
color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('Bike Rides', color=color)
ax1.plot(df_cleaned['date'], df_cleaned['bike_rides_daily'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Avg Temperature (C)', color=color)  
ax2.plot(df_cleaned['date'], df_cleaned['avgTemp'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  
plt.title('Daily Bike Rides and Avg Temperature (2022)')
plt.show()
```
