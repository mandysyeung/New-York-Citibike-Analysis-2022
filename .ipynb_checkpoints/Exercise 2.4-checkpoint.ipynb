```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "C:/Users/msyeu/New-York-Citibike-Analysis-2022/merged_citibike_weather.csv"
```

```python
dtype_spec = {
    'start_station_id': str,
    'end_station_id': str
}
```

```python
chunk_size = 100000
chunks = []
```

```python
for chunk in pd.read_csv(file_path, dtype=dtype_spec, chunksize=chunk_size, low_memory=False):
    # Take a 25% sample from each chunk
    chunks.append(chunk.sample(frac=0.25, random_state=42))
```

```python
df_bike = pd.concat(chunks)
```

```python
print("Subset size:", df_bike.shape)
```

```python
df_bike.head()
```

```python
df_top_stations = df_bike.groupby('start_station_name')['ride_id'].count().reset_index()
```

```python
df_top_stations.columns = ['start_station_name', 'ride_count']
```

```python
# Selecting the top 20 most frequent start stations
df_top20_stations = df_top_stations.nlargest(20, 'ride_count')
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=df_top20_stations, x='ride_count', y='start_station_name', palette='Blues_r')
plt.title("Top 20 Starting Stations by Frequency")
plt.xlabel("Ride Count")
plt.ylabel("Starting Station")
plt.show()
```

```python
df_bike['tripduration_minutes'] = (df_bike['ended_at'] - df_bike['started_at']).dt.total_seconds() / 60
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='rideable_type', y='tripduration_minutes', data=df_bike)
plt.title('Box Plot of Rideable Type vs. Trip Duration')
plt.ylabel('Trip Duration (minutes)')
plt.xlabel('Rideable Type')
plt.show()
```

A `FacetGrid` can be a powerful tool to visualize relationships between multiple variables across different subsets of the data. For instance, you might want to analyze how the trip duration varies by both `rideable_type` and whether the user is a `member` or `casual` rider. This would allow you to compare the trip durations across different user types and bike types, which might help explain patterns in bike usage behavior.


```python
df_bike['tripduration_minutes'] = (df_bike['ended_at'] - df_bike['started_at']).dt.total_seconds() / 60
```

```python
import numpy as np
```

```python
df_bike.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df_bike_filtered = df_bike[(df_bike['tripduration_minutes'] > 1) & (df_bike['tripduration_minutes'] <= 120)]
```

```python
grid = sns.FacetGrid(df_bike_filtered, col="rideable_type", row="member_casual", margin_titles=True)
grid.map(sns.histplot, "tripduration_minutes", bins=15)
grid.set(xlim=(0, 120))
```

The FacetGrid plot reveals that for both classic bikes and electric bikes, the majority of trip durations tend to be shorter, especially within the first 20 minutes. The distributions show a sharp decline in frequency as trip duration increases, indicating that users tend to prefer shorter trips. Additionally, casual riders, particularly with electric bikes, seem to have a slightly longer average trip duration compared to members, suggesting that casual users might use the service more for leisure, whereas members might use it for shorter, more routine trips.
